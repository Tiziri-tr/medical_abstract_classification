```python
import pickle
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

df = pickle.load(open('/home/animal/PycharmProjects/creative_thinking/dataset/preprocessed_eng_dataset.pickle', 'rb'))
```

```python
def plot_bar_x(labels, values):
    # this is for plotting purpose
    index = np.arange(len(labels))
    plt.bar(index, values)
    plt.xlabel('Genre', fontsize=5)
    plt.ylabel('No of Movies', fontsize=5)
    plt.xticks(index, labels, fontsize=5, rotation=30)
    plt.title('Tags distributions')
    plt.show()
```

```python
l = Counter(df['tag'])
labels = l.keys()
values = list(l.values())
plot_bar_x(labels, values)
```

```python
from yellowbrick.text import TSNEVisualizer
from sklearn.feature_extraction.text import TfidfVectorizer

# load data
#corpus = load_corpus('hobbies')
tfidf  = TfidfVectorizer()
corpus = df.loc[df['language'] == "eng"]
abstr   = tfidf.fit_transform(corpus["abstract"])


# Create the visualizer
tsne = TSNEVisualizer()
tsne.fit(abstr, corpus["tag"])
tsne.poof()
```

```python
from sklearn.cluster import KMeans

#corpus = df["abstract"].loc[df['language'] == "eng"]
clusters = KMeans(n_clusters=39)
clusters.fit(abstr)

```

```python
tsne = TSNEVisualizer()
tsne.fit(abstr, ["c{}".format(c) for c in clusters.labels_])
tsne.poof()
```
